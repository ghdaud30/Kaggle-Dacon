```python
def item_code_by_item_name(item_name):
  item_code_list = stocks.loc[stocks['Name'] == item_name,'Code'].tolist()
  if(len(item_code_list) > 0):
    return item_code_list[0]
  else:
    return False

item_code_by_item_name('하이브')
```

```python
item_code_by_item_name('엘지전자')
```

```python
item_code_by_item_name('LG전자')
```

```python
fdr.DataReader(item_code_by_item_name('하이브'),'2022')
```

```python
def find_item_list(item_name, year = 2020):
  item_code = item_code_by_item_name(item_name)
  if item_code:
    df_day = fdr.DataReader(item_code,str(year))
    return df_day
  else:
    return False

find_item_list('하이브',2022)
```

```python
stock_daily = find_item_list('하이브',2022)
stock_daily.plot(subplots=True,figsize=(15,10))
```

```python
stock_daily[['Close','Volume']].plot(secondary_y='Volume')
```

```python
df = fdr.DataReader('005930','2020','2022')
df
```

#전일 대비 수익률 구하기
- Change = (종가 / 전일종가) - 1

```python
df['Change2'] = df['Close'] / df['Close'].shift(1) - 1
df[['Change','Change2']]
```

```python
df['pct_change'] = df['Close'].pct_change()
df[['Change','Change2','pct_change']]
```

```python
df['Close'].diff()
```

#Rolling

```python
df['SMA7'] =  df["Close"].rolling(7).mean()
df['SMA14'] = df["Close"].rolling(14).mean()
df[['Close','SMA7','SMA14']].plot()
```

```python
df['CMA'] = df['Close'].expanding(14).mean()
df['CMA']
```

```python
df[['Close','SMA7','SMA14','CMA']].plot()
```

```python
df['EMW_long'] = df['Close'].ewm(26).mean()
df['EMW_long']
```

```python
df[['Close','EMW_long']].plot()
```
