```python
df['income_cat'] = np.ceil(df['median_income'] / 1.5)
df['income_cat'].where(df['income_cat'] < 5, other=5.0,inplace=True)
df['income_cat']
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
split = StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=42)
```

```python
for train_index , test_index in split.split(df,df['income_cat']):
  strat_train_set = df.loc[train_index]
  strat_test_set = df.loc[test_index]
```

```python
from sklearn.model_selection import train_test_split

train_set , test_set = train_test_split(df,test_size=0.2, random_state = 42)

def income_cat_prop(data):
  return data["income_cat"].value_counts() / len(data)

income_cat_prop(df)
```

```python
comparing_prop = pd.DataFrame({
    "Overall" : income_cat_prop(df),
    "Random" : income_cat_prop(test_set),
    "Strat" : income_cat_prop(strat_train_set)
}).sort_index()
```

```python
comparing_prop['Random % error'] = 100 * comparing_prop['Random'] / comparing_prop['Overall'] - 100
comparing_prop['Strat % error'] = 100 * comparing_prop['Strat'] / comparing_prop['Overall'] - 100
comparing_prop
```
