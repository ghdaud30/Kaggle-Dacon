```python
import sys
import pymysql

def mysqlConnection(user,password,host,port,database):
    try:
        con = pymysql.connect(user=user,password=password,host=host,port=port,database=database)
        print('DBConnection is completed!')
    except pymysql.Error as e:
        print(f'DBConnection error : {e}')
        sys.exit(1)
        
    return con

def mysqlDbClose(corn):
    try:
        corn.close()
        print("DB Close Success")
    except pymysql.Error as e:
        print(f'DB Close is Error : {e}')
        sys.exit(1)
```

```python
dbCon = mysqlConnection('multicultural', 'abcd', '127.0.0.1', 3306, 'multicultural')
cursor = dbCon.cursor()

file = open('./data/다문화가족.csv',encoding='UTF-8')
fReader = csv.reader(file)

# 첫 번째 줄은 헤더로 처리하지 않고 데이터로 읽기
header_skipped = False
for line in fReader:
    if not header_skipped:
        header_skipped = True
        continue  # 헤더를 스킵함
    print(line)
    query = "INSERT INTO multicultural VALUES ('{0}', '{1}', '{2}', '{3}', '{4}', '{5}','{6}')".format(line[0], line[1], line[2], line[3],line[4],line[5],line[6])
    cursor.execute(query)

# 변경사항 커밋 
dbCon.commit() 
# 커넥션 닫기
mysqlDbClose(dbCon)
```
